```python
#function to get data columns and put them into lists
def get_data(col):
    #data for x column
    data = [
            df1L7[col],
            df2L7[col],
            df3L7[col],
            df4L7[col],
            df5L7[col],
            df1L6[col],
            df2L6[col],
            df3L6[col],
            df4L6[col],
            df5L6[col],
            df1L5[col],
            df2L5[col],
            df3L5[col],
            df4L5[col],
            df5L5[col],
            df1L4[col],
            df2L4[col],
            df3L4[col],
            df4L4[col],
            df5L4[col],
            df1L3[col],
            df2L3[col],
            df3L3[col],
            df4L3[col],
            df5L3[col]
            ]
    return data

loss = get_data("loss")
mae = get_data("mae")
mean_q = get_data("mean_q")
mean_eps = get_data("mean_eps")
episode_reward = get_data("episode_reward")
nb_episode_steps = get_data("nb_episode_steps")
nb_steps = get_data("nb_steps")
episode = get_data("episode")
duration = get_data("duration")

cols = ["loss", "mae", "mean_q", "mean_eps", "episode_reward", "nb_episode_steps", "nb_steps", "episode", "duration"]
```
