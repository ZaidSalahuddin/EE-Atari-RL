```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path7 = r"7 Layers\\"
df1L7 = pd.read_csv(path7+r"Cone\\7L_Cone.csv")
df2L7 = pd.read_csv(path7+r"ConeInv\\7L_ConeInv.csv")
df3L7 = pd.read_csv(path7+r"HourG\\7L_HourG.csv")
df4L7 = pd.read_csv(path7+r"HourGInv\\7L_HourGInv.csv")
df5L7 = pd.read_csv(path7+r"Line\\7L_Line.csv")

path6 = r"6 Layers\\"
df1L6 = pd.read_csv(path6+r"Cone\\6L_Cone.csv")
df2L6 = pd.read_csv(path6+r"ConeInv\\6L_ConeInv.csv")
df3L6 = pd.read_csv(path6+r"HourG\\6L_HourG.csv")
df4L6 = pd.read_csv(path6+r"HourGInv\\6L_HourGInv.csv")
df5L6 = pd.read_csv(path6+r"Line\\6L_Line.csv")

path5 = r"5 Layers\\"
df1L5 = pd.read_csv(path5+r"Cone\\5L_Cone.csv")
df2L5 = pd.read_csv(path5+r"ConeInv\\5L_ConeInv.csv")
df3L5 = pd.read_csv(path5+r"HourG\\5L_HourG.csv")
df4L5 = pd.read_csv(path5+r"HourGInv\\5L_HourGInv.csv")
df5L5 = pd.read_csv(path5+r"Line\\5L_Line.csv")

path4 = r"4 Layers\\"
df1L4 = pd.read_csv(path4+r"Cone\\4L_Cone.csv")
df2L4 = pd.read_csv(path4+r"ConeInv\\4L_ConeInv.csv")
df3L4 = pd.read_csv(path4+r"HourG\\4L_HourG.csv")
df4L4 = pd.read_csv(path4+r"HourGInv\\4L_HourGInv.csv")
df5L4 = pd.read_csv(path4+r"Line\\4L_Line.csv")

path3 = r"3 Layers\\"
df1L3 = pd.read_csv(path3+r"Cone\\3L_Cone.csv")
df2L3 = pd.read_csv(path3+r"ConeInv\\3L_ConeInv.csv")
df3L3 = pd.read_csv(path3+r"HourG\\3L_HourG.csv")
df4L3 = pd.read_csv(path3+r"HourGInv\\3L_HourGInv.csv")
df5L3 = pd.read_csv(path3+r"Line\\3L_Line.csv")
```

```python
#function to get data columns and put them into lists
def get_data(col):
    #data for x column
    data = [
            df1L7[col],
            df2L7[col],
            df3L7[col],
            df4L7[col],
            df5L7[col],
            df1L6[col],
            df2L6[col],
            df3L6[col],
            df4L6[col],
            df5L6[col],
            df1L5[col],
            df2L5[col],
            df3L5[col],
            df4L5[col],
            df5L5[col],
            df1L4[col],
            df2L4[col],
            df3L4[col],
            df4L4[col],
            df5L4[col],
            df1L3[col],
            df2L3[col],
            df3L3[col],
            df4L3[col],
            df5L3[col]
            ]
    return data

loss = get_data("loss")
mae = get_data("mae")
mean_q = get_data("mean_q")
mean_eps = get_data("mean_eps")
episode_reward = get_data("episode_reward")
nb_episode_steps = get_data("nb_episode_steps")
nb_steps = get_data("nb_steps")
episode = get_data("episode")
duration = get_data("duration")

cols = ["loss", "mae", "mean_q", "mean_eps", "episode_reward", "nb_episode_steps", "nb_steps", "episode", "duration"]
```

```python
#subplot one
alpha = 1
lw = 0.7
y_lim_min = -10
y_lim_max = 140

#set up the plot
fig = plt.figure(figsize=(20,12))
ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

fig.suptitle('Mean Q by Layers', fontsize=30)

#plotting the lines
#7 layer
label = ["7 Cone","7 Cone Inv","7 HourGlass","7 HourGlass Inv","7 Line"]
color = ["red","green","blue","black","orange"]
for i in range(5):
    ax1.plot(episode[i], mean_q[i], alpha=alpha, label = label[int(i)], lw=lw, color=color[int(i)])
ax1.title.set_text('7 Layers')
ax1.set_xlabel('episode')
ax1.set_ylabel('mean_q')
ax1.legend(loc=0)
ax1.set_ylim([y_lim_min,y_lim_max])

#6 layer
label = ["6 Cone","6 Cone Inv","6 HourGlass","6 HourGlass Inv","6 Line"]
color = ["red","green","blue","black","orange"]
for i in range(5):
    ax2.plot(episode[i+5], mean_q[i+5], alpha=alpha, label = label[int(i)], lw=lw, color=color[int(i)])
ax2.title.set_text('6 Layers')
ax2.set_xlabel('episode')
ax2.set_ylabel('mean_q')
ax2.legend(loc=0)
ax2.set_ylim([y_lim_min,y_lim_max])

#5 layer
label = ["5 Cone","5 Cone Inv","5 HourGlass","5 HourGlass Inv","5 Line"]
color = ["red","green","blue","black","orange"]
for i in range(5):
    ax3.plot(episode[i+10], mean_q[i+10], alpha=alpha, label = label[int(i)], lw=lw, color=color[int(i)])
ax3.title.set_text('5 Layers')
ax3.set_xlabel('episode')
ax3.set_ylabel('mean_q')
ax3.legend(loc=0)
ax3.set_ylim([y_lim_min,y_lim_max])

#4 layer
label = ["4 Cone","4 Cone Inv","4 HourGlass","4 HourGlass Inv","4 Line"]
color = ["red","green","blue","black","orange"]
for i in range(5):
    ax4.plot(episode[i+15], mean_q[i+15], alpha=alpha, label = label[int(i)], lw=lw, color=color[int(i)])
ax4.title.set_text('4 Layers')
ax4.set_xlabel('episode')
ax4.set_ylabel('mean_q')
ax4.legend(loc=0)
ax4.set_ylim([y_lim_min,y_lim_max])

#3 layer
label = ["3 Cone","3 Cone Inv","3 HourGlass","3 HourGlass Inv","3 Line"]
color = ["red","green","blue","black","orange"]
for i in range(5):
    ax5.plot(episode[i+20], mean_q[i+20], alpha=alpha, label = label[int(i)], lw=lw, color=color[int(i)])
ax5.title.set_text('3 Layers')
ax1.set_xlabel('episode')
ax5.set_ylabel('mean_q')
ax5.legend(loc=0)
ax5.set_ylim([y_lim_min,y_lim_max])
```
