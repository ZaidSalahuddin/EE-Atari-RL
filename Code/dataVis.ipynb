```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"test\test.csv")
df2 = pd.read_csv(r"test 2\test2.csv")
df3 = pd.read_csv(r"test 3\test3.csv")
df4 = pd.read_csv(r"test 4\test4.csv")
df5 = pd.read_csv(r"test 5\test5.csv")
df6 = pd.read_csv(r"test 6\test6.csv")
```

```python
cols = ["loss", "mae", "mean_q", "mean_eps", "episode_reward", "nb_episode_steps", "nb_steps", "episode", "duration"]

tCol = 'mean_q'
tX = 'episode'

x1 = df[tX]
x2 = df2[tX]
x3 = df3[tX]
x4 = df4[tX]
x5 = df5[tX]
x6 = df6[tX]

y1 = df[tCol]
y2 = df2[tCol]
y3 = df3[tCol]
y4 = df4[tCol]
y5 = df5[tCol]
y6 = df6[tCol]

alpha = 0.7
lw = 0.7
 
fig = plt.figure(figsize=(20,12))
ax = fig.add_axes([0,0,1,1])

ax.plot(x1,y1, alpha=alpha, label = "test 1", lw=lw)
ax.plot(x2,y2, alpha=alpha, label = "test 2", lw=lw)
ax.plot(x3,y3, alpha=alpha, label = "test 3", lw=lw)
ax.plot(x4,y4, alpha=alpha, label = "test 4", lw=lw)
ax.plot(x5,y5, alpha=alpha, label = "test 5", lw=lw)
ax.plot(x6,y6, alpha=alpha, label = "test 6", lw=lw)

ax.set_xlabel(tX)
ax.set_ylabel(tCol)
ax.legend(loc=0)
```
