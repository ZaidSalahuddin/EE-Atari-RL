```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"7 Layers\\"
df1 = pd.read_csv(path+r"Cone\\7L_Cone.csv")
df2 = pd.read_csv(path+r"ConeInv\\7L_ConeInv.csv")
df3 = pd.read_csv(path+r"HourG\\7L_HourG.csv")
df4 = pd.read_csv(path+r"HourGInv\\7L_HourGInv.csv")
df5 = pd.read_csv(path+r"Line\\7L_Line.csv")
```

```python
cols = ["loss", "mae", "mean_q", "mean_eps", "episode_reward", "nb_episode_steps", "nb_steps", "episode", "duration"]

tCol = 'mean_q'
tX = 'episode'

x1 = df1[tX]
y1 = df1[tCol]

x2 = df2[tX]
y2 = df2[tCol]

x3 = df3[tX]
y3 = df3[tCol]

x4 = df4[tX]
y4 = df4[tCol]

x5 = df5[tX]
y5 = df5[tCol]

alpha = 0.7
lw = 0.7
 
fig = plt.figure(figsize=(20,12))
ax = fig.add_axes([0,0,1,1])

ax.plot(x1,y1, alpha=alpha, label = "Cone", lw=lw)
ax.plot(x2,y2, alpha=alpha, label = "Cone Inv", lw=lw)
ax.plot(x3,y3, alpha=alpha, label = "HourG Inv", lw=lw)
ax.plot(x4,y4, alpha=alpha, label = "HourG Inv", lw=lw)
ax.plot(x5,y5, alpha=alpha, label = "Line", lw=lw)

ax.set_xlabel(tX)
ax.set_ylabel(tCol)
ax.legend(loc=0)
```
