```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"test\test_1_4win.csv")
df2 = pd.read_csv(r"test\test_01.csv")
df3 = pd.read_csv(r"test\test_1.csv")
df4 = pd.read_csv(r"test\test_other1.csv")
```

```python
cols = ["loss", "mae", "mean_q", "mean_eps", "episode_reward", "nb_episode_steps", "nb_steps", "episode", "duration"]

tCol = 'episode_reward'
tX = 'episode'

x1 = df[tX]
y1 = df[tCol]

x2 = df2[tX]
y2 = df2[tCol]

x3 = df3[tX]
y3 = df3[tCol]

x4 = df4[tX]
y4 = df4[tCol]

alpha = 0.7
lw = 0.7
 
fig = plt.figure(figsize=(20,12))
ax = fig.add_axes([0,0,1,1])

ax.plot(x1,y1, alpha=alpha, label = "test 4win", lw=lw)
ax.plot(x2,y2, alpha=alpha, label = "test 1", lw=lw)
ax.plot(x3,y3, alpha=alpha, label = "idk", lw=lw)
ax.plot(x4,y4, alpha=alpha, label = "idk", lw=lw)


ax.set_xlabel(tX)
ax.set_ylabel(tCol)
ax.legend(loc=0)
```
